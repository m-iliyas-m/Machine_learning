# Bivariate and Multivariate Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
df = pd.read_csv('C:\\Users\\hp\\Desktop\\data\\tips.csv')
df1 = pd.read_csv('C:\\Users\\hp\\Desktop\\data\\titanic.csv')
df2= sns.load_dataset('flights')
df3 = sns.load_dataset('iris')
```

```python
df.head()
```

## 1. Scatterplot( Numerical - Numerical)

```python
sns.scatterplot(df['total_bill'],df['tip'],hue=df['sex'],style = df['smoker'],size=df['size'])
```

## 2. Bar plot (Numerical - Categorical)

```python
sns.barplot(df1['Pclass'],df1['Age'],hue=df1['Sex'])
```

## 3. boxplot (Numerical -Categorical)

```python
sns.boxplot(df1['Sex'],df1['Age'],hue=df1['Embarked'])
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(df1[df1['Survived']==0]['Age'],hist=False)
sns.distplot(df1[df1['Survived']==1]['Age'],hist=False)
```

## 5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(df1['Pclass'],df1['Survived']))
```

```python
(df1.groupby('Embarked').mean()['Survived']*100).plot(kind='bar')
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(df1['Parch'],df1['Survived']))
```

## 7. Pairplot

```python
sns.pairplot(df3 , hue= 'species')
```

## 8. lineplot (numerical - numerical)

```python
new = df2.groupby('year').sum().reset_index()
new.head(1)
```

```python
sns.lineplot(new["year"],new['passengers'])
```

```python
sns.heatmap(df2.pivot_table(values = 'passengers',index='month',columns = 'year'))
```

```python
sns.clustermap(df2.pivot_table(values = 'passengers',index='month',columns = 'year'))
```
